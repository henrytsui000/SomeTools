# Compile and Calculate Time

```python
import os
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
Sort_Min, Sort_Max = 1, 100000
Simu_list = [2**i for i in range(int(math.log2(Sort_Min)), round(math.log2(Sort_Max))+1)]
print(Simu_list)
os.system("g++ ./ans.cpp")
for num in Simu_list:
    os.system(f"./a.out {num} > ./out/{num}")
```

# Statistic and Analysis

```python
tmp = os.listdir("./out")
file_name = [int(x) for x in tmp]
file_name.sort()
print(file_name)
```

```python
All_result = []
for name in file_name:
    tmp = []
    f = open(f"./out/{name}", "r")
    tp = f.readlines()
    for line in tp:
        tmp.append(line.split(' ')[-1][:-1])
    All_result.append(tmp)
print(All_result)
print(All_result[0][:])
```

```python
dta= np.array(All_result, dtype="float")
```

```python
plt.title("Two Sort Algorithm data")
plt.ylabel("Using Time")
plt.xlabel("Data Size")
plt.plot(dta[:,0], dta[:,2], 'r^-', label = "Insertion Sort")
plt.plot(dta[:,0], dta[:,5], 'b^-', label = "Merge Sort")
plt.legend()
plt.show()
plt.title("Insertoin Sort's Using time : data / size")
plt.ylabel("Using Time")
plt.xlabel("Data Size")
plt.plot(dta[:,0], dta[:,2]/dta[:,0],'ro-' ,label = "Insertion Sort")
# plt.plot(dta[:,0], dta[:,5]/dta[:,0],'bo-' ,label = "Merge Sort")
# plt.xscale("log", basex=2)
plt.legend()
plt.show()

plt.title("Merge Sort's Using time : data / size")
plt.ylabel("Using Time")
plt.xlabel("Data Size")
# plt.plot(dta[:,0], dta[:,2]/dta[:,0],'ro-' ,label = "Insertion Sort")
plt.plot(dta[:,0], dta[:,5]/dta[:,0],'bo-' ,label = "Merge Sort")
plt.xscale("log", basex=2)
plt.legend()
plt.show()
```
