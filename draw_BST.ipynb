```python
import matplotlib.pyplot as plt
import networkx as nx
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
T = nx.Graph()
f = open("tree.txt")
lines = f.readlines()
for line in lines:
  T.add_edge(*line.split())
f.close()
```

```python
pos = graphviz_layout(T, prog="dot")
nx.draw(T,pos, with_labels=True)
plt.show()
```
